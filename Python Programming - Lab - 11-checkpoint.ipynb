### 01) WAP to demonstrate the use of Pie chart.

```python
price=[2000,5000,3500,1000,10000]
material=['silver','mercury','aluminum','coper','gold']
c=['r','g','b','c','m']
plt.pie(price,colors=c,labels=material,shadow=True,autopct='%1.1f%%')
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x=[1,2,3,4,5]
lis=['st1','st2','st3','st4','st5']
y=[50,45,38,49,25]
c=['r','g','b','c','m']
plt.title('python marks')
plt.bar(x,y,color=c,label=lis)
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=[1,2,3,4,5]
lis=['st1','st2','st3','st4','st5']
y=[50,45,38,49,25]
c=['r','g','b','c','m']
plt.title('python marks')
plt.bar(x,y,color=c,label=lis)
for i in range(1,len(x)+1):
    plt.text(i,y[i-1],lis[i-1])
plt.show()
```
